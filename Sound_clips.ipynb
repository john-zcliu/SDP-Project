```python
array = []
key_words = []
for d in range(len(clips)):
    print("iteration")
    if(d!=len(array)):
        print(d)
        print("problem")
        break
    xa= clips["Description"].iloc[d].lower()
    words = [w for w in nltk.word_tokenize(xa)]
    found = False
    for a in range(len(nltk.pos_tag(words))):
        x = nltk.pos_tag(words)[a][0].lower()
        y = nltk.pos_tag(words)[a][1]
        
        if(y == "JJ" and "wood" in x):
            key_words.append(x)
            array.append("wood")
            found = True
            break
        
        elif ((y == "NN" or y == "NNS") and ("bunch" not in x and "set" not in x and "pile" not in x and "medium" not in x
                                          and "spreader" not in x and "movement" not in x and "standard" not in x
                                          and "chunk" not in x and "element" not in x and "toonie" not in x
                                          and "handful" not in x)):
            print(x)
            print(y)
            key_words.append(x)
            if("plastic" in x or "botte" in x):
                array.append("plastic")
            elif ("key" in x or "keys" in x or "metal" in x or "aluminum" in x or "coin" in x or "silverware" in x):
                array.append("metal")
            elif ("cardboard" in x or "storage" in x or "box" in x):
                array.append("cardboard")
            elif ("glass" in x):
                array.append("glass")
            elif ("paper" in x):
                array.append("paper")
            elif ("wood" in x.lower()):
                array.append("wood")
            else:
                array.append("trash")
            found = True
            break
        
    if(not found):
        for a in range(len(nltk.pos_tag(words))):
            x = nltk.pos_tag(words)[a][0]
            y = nltk.pos_tag(words)[a][1]
            if (y == "JJ"):
                key_words.append(x)
                array.append("trash")
                break
            
    
```

```python
def get_window(filename, sr=22050, wnd_size=2000):
    x, _ = librosa.load(filename, sr=22050)
    y = np.cumsum(x**2)
    y /= y[-1]
    z = y[wnd_size:] - y[:-wnd_size]
    s = z.argmax()
    e = s + wnd_size
    return x[s:e]
```

```python
clips_arrays = []
tags = []
for x in clips["clips"]:
    print(x)
    try:
        clips_arrays.append(get_window("../clips/"+x))
        tags.append(x)
    except:
        continue
```

```python
clips = clips[clips["clips"].map(set(tags).__contains__)]
```
